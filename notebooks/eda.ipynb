```python
import pandas as pd
risk_data = pd.read_csv(r'C:\Users\HP\credit-risk-model\data\raw\data.csv')
print(risk_data.head())
```

```python
risk_data.info()
```

```python
id_columns = [col for col in risk_data.columns if col.lower().endswith('id')]
print(id_columns)
```

```python
# Creating a sample dataframe that matches your data
# Extract the number after the underscore and convert it to an integer
risk_data[id_columns] = risk_data[id_columns].apply(lambda x: x.str.split('_').str[1].astype(int))

print("\nAfter transformation:")
print(risk_data.head())
print(risk_data.dtypes)
```

```python
date_cols = ['TransactionStartTime']
risk_data[date_cols] = risk_data[date_cols].apply(pd.to_datetime)

print(risk_data[['TransactionStartTime']].dtypes)
print(risk_data[['TransactionStartTime']].head())
```
